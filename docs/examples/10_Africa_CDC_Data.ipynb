# 🌍 Africa CDC Data Access

This notebook demonstrates how to access public health surveillance data from the **African Centres for Disease Control and Prevention (Africa CDC)**.

## Overview

The Africa CDC coordinates disease surveillance and response across the African continent, covering:
- **Disease outbreak surveillance**: Weekly outbreak briefs
- **COVID-19 data**: Historical pandemic surveillance
- **Event-Based Surveillance (EBS)**: Real-time threat detection
- **Vaccination coverage**: Immunization program data
- **Priority diseases**: Ebola, Marburg, Lassa fever, Cholera, Mpox, etc.

**Website:** https://africacdc.org/

**Coverage:** 55 African Union Member States

## 📦 Setup

```python
import sys

import matplotlib.pyplot as plt
import pandas as pd

from epidatasets.sources.africa_cdc import AfricaCDCAccessor

# Initialize accessor
africa_cdc = AfricaCDCAccessor()

print("✅ Africa CDC Accessor initialized successfully!")
```

## 🌍 African Union Member States

Africa CDC covers all 55 African Union member states:

```python
# List all countries
countries = africa_cdc.list_countries()
print(f"Total AU member states: {len(countries)}")
print("\nFirst 15 countries:")
countries.head(15)
```

## 🗺️ Africa CDC Regions

Africa CDC organizes countries into 5 geographical regions:

```python
# View countries by region
regions = africa_cdc.list_regions()

for region in regions['region'].unique():
    region_countries = regions[regions['region'] == region]
    print(f"\n📍 {region} Region ({len(region_countries)} countries):")
    country_names = region_countries['country_name'].tolist()
    print("  " + ", ".join(country_names[:8]))
    if len(country_names) > 8:
        print(f"  ... and {len(country_names) - 8} more")
```

## 🦠 Priority Diseases

Africa CDC tracks priority diseases for the continent:

```python
# List priority diseases
diseases = africa_cdc.list_priority_diseases()
print(f"Total priority diseases: {len(diseases)}\n")
diseases
```

```python
# Categorize diseases by type
hemorrhagic_fevers = ['EBOLA', 'MARBRG', 'LASSA', 'CCHF']
vbd = ['DENGUE', 'MALARIA', 'YELLOW_FEVER', 'RVF']  # Vector-borne
vaccine_preventable = ['MEASLES', 'MENINGITIS', 'INFLUENZA', 'YELLOW_FEVER']

print("🩸 Viral Hemorrhagic Fevers:")
for d in hemorrhagic_fevers:
    print(f"  - {d}: {africa_cdc.PRIORITY_DISEASES[d]}")

print("\n🦟 Vector-Borne Diseases:")
for d in vbd:
    print(f"  - {d}: {africa_cdc.PRIORITY_DISEASES[d]}")

print("\n💉 Vaccine-Preventable Diseases:")
for d in vaccine_preventable:
    print(f"  - {d}: {africa_cdc.PRIORITY_DISEASES[d]}")
```

## 📊 Disease Outbreak Surveillance

Access current and historical disease outbreak data:

```python
# Get outbreak data for Ebola
ebola_outbreaks = africa_cdc.get_disease_outbreaks(disease='EBOLA')

print(f"Ebola outbreak records: {len(ebola_outbreaks)}")
print("\nSample structure:")
ebola_outbreaks.head()
```

```python
# Get outbreaks in East Africa
east_africa = africa_cdc.get_countries_by_region('Eastern')
print(f"East African countries: {len(east_africa)}")

east_africa_outbreaks = africa_cdc.get_disease_outbreaks(
    countries=east_africa[:10]  # First 10 for demo
)
print(f"\nOutbreak records for East Africa sample: {len(east_africa_outbreaks)}")
```

## 🦠 COVID-19 Historical Data

Access COVID-19 surveillance data for African countries:

```python
# Get COVID-19 data for selected countries
covid_data = africa_cdc.get_covid_data(
    countries=['ZA', 'NG', 'ET', 'KE', 'GH'],  # 5 major economies
    date_range=('2020-03-01', '2021-12-31')
)

print(f"COVID-19 records: {len(covid_data)}")
print("\nCountries included:")
covid_data[['country_code', 'country_name']].drop_duplicates()
```

## 💉 Vaccination Coverage

Access vaccination coverage data:

```python
# Get vaccination data for selected countries
vax_data = africa_cdc.get_vaccination_coverage(
    countries=['NG', 'ET', 'ZA', 'KE'],
    vaccines=['COVID-19', 'Measles']
)

print(f"Vaccination records: {len(vax_data)}")
vax_data.head(10)
```

## 🚨 Event-Based Surveillance (EBS)

Access real-time event-based surveillance alerts:

```python
# Get EBS alerts for Sahel region
sahel_countries = ['ML', 'NE', 'TD', 'BF', 'MR']

ebs_alerts = africa_cdc.get_event_based_surveillance(
    countries=sahel_countries,
    date_range=('2024-01-01', '2024-12-31')
)

print(f"EBS alerts for Sahel: {len(ebs_alerts)}")
ebs_alerts.head()
```

## 📋 Weekly Outbreak Briefs

Africa CDC publishes weekly outbreak briefs:

```python
# Get outbreak briefs for a year
briefs_2024 = africa_cdc.get_weekly_outbreak_brief(year=2024)

print(f"Total briefs in 2024: {len(briefs_2024)}")
print("\nFirst 5 weeks:")
briefs_2024.head()[['year', 'week', 'data_source']]
```

```python
# Get specific week
week_30 = africa_cdc.get_weekly_outbreak_brief(year=2024, week=30)
print("Week 30 brief:")
week_30[['year', 'week', 'diseases_covered']]
```

## 📈 Regional Comparison

Compare disease occurrence across regions:

```python
# Compare COVID-19 across regions
regional_comparison = africa_cdc.compare_regions(
    disease='COVID-19',
    year=2021
)

print("Regional comparison structure:")
regional_comparison if not regional_comparison.empty else "No data available"
```

## 📊 Country Summary

Get outbreak summary by country:

```python
# Get summary for all diseases
summary_2024 = africa_cdc.get_summary_by_country(year=2024)

print(f"Summary records: {len(summary_2024)}")
summary_2024.head(10) if not summary_2024.empty else "No data available"
```
